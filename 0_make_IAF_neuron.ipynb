```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
import nengo
```

Input = Sine wave

Hidden layer = Nengo.ensemble: all to all network


Input --> synapse (0.1) --> Hidden layer

```
with nengo.Network() as net:
    inp = nengo.Node(output=np.sin)
    ens = nengo.Ensemble(1, 1, neuron_type=nengo.LIF(), intercepts=[0])
    # n_neurons, dimensions, neuron_type, intercepts 0 - 1
    nengo.Connection(inp, ens, synapse=0.1)

    p = nengo.Probe(ens)
    inp_p = nengo.Probe(inp)
    spike_p = nengo.Probe(ens.neurons)
    voltage_p = nengo.Probe(ens.neurons, 'voltage')
```

```
with nengo.Simulator(net) as sim:
    plt.figure()
    plt.plot(*nengo.utils.ensemble.tuning_curves(ens, sim))
    plt.xlabel("input value")
    plt.ylabel("firing rate")
    plt.title(str(nengo.LIF()))
    sim.run(5.0)   # 5s
```

```
plt.figure()
 
plt.plot(sim.trange(), sim.data[p])
plt.plot(sim.trange(), sim.data[inp_p])
```

```
plt.figure()
 
plt.plot(sim.trange(), sim.data[voltage_p])
plt.plot(sim.trange(), sim.data[inp_p])
```
